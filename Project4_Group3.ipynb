```python
import os
import xlsxwriter

class Process_Text_Data:
    def __init__(self):
        pass
    
    def extract_data_into_lists(self):
        RAW_FILE_NAME = "APL LiveChat Data Set by StartEndTimeChat 2021.txt"
        FILE_NAME = os.path.join(os.path.abspath(''), RAW_FILE_NAME)
        
        with open(FILE_NAME, 'r') as input_file:
            all_data = input_file.read()
            
            all_started_ended_lines = []
            self.get_all_started_ended_lines(all_data, all_started_ended_lines)
            
            # Do not delete this comment line
            #not_completed_get_all_started_ended_lines(self, all_data, all_started_ended_lines)
            
            filtered_started_ended_lines = []
            self.get_filtered_started_ended_lines(all_started_ended_lines, filtered_started_ended_lines)
            
            all_year_date_started_ended_lines = []
            self.get_year_date_started_ended_lines(filtered_started_ended_lines, all_year_date_started_ended_lines)
            
            all_year_month_date_started_ended_lines = []
            self.get_year_month_date_started_ended_lines(all_year_date_started_ended_lines, all_year_month_date_started_ended_lines)
            
            self.export_to_excel(all_year_month_date_started_ended_lines)
    
    def get_all_started_ended_lines(self, all_data, all_started_ended_lines):
        raw_all_started_ended_lines = all_data.split(',')

        tmp_all_started_ended_lines = []
        for raw_started_ended_line in raw_all_started_ended_lines:
            tmp_all_started_ended_lines.append(raw_started_ended_line.strip())
            
        index = 0
        while index < len(tmp_all_started_ended_lines):
            if (index + 1 <= len(tmp_all_started_ended_lines)) and ('\"started_at\":' in tmp_all_started_ended_lines[index]) and ('\"ended_at\":' in tmp_all_started_ended_lines[index + 1]):
                all_started_ended_lines.append(
                    tmp_all_started_ended_lines[index] +  ' ' + tmp_all_started_ended_lines[index + 1])
                index = index + 2
                
            else:
                index = index + 1
        
        # For debugging purposes
        #self.print_list(all_started_ended_lines)
    
    # This method is for debugging purposes.
    # Do not delete this method
    def not_completed_get_all_started_ended_lines(self, all_data, all_started_ended_lines):
        raw_all_started_ended_lines = all_data.split(',')

        tmp_all_started_ended_lines = []
        for raw_started_ended_line in raw_all_started_ended_lines:
            tmp_all_started_ended_lines.append(raw_started_ended_line.strip())
            
        index = 0
        while index < len(tmp_all_started_ended_lines):
            all_started_ended_lines.append(
                tmp_all_started_ended_lines[index] +  ' ' + tmp_all_started_ended_lines[index + 1])
            index = index + 2
            
        # For debugging porposes
        #self.print_list(all_started_ended_lines)
            
    def get_filtered_started_ended_lines(self, all_started_ended_lines, filtered_started_ended_lines):
        for started_ended_line in all_started_ended_lines:
            tokens = started_ended_line.split(' ')
            filtered_started_ended_lines.append(tokens[1] + ' ' + tokens[3])
        
        # For debugging porposes
        #self.print_list(filtered_started_ended_lines)
        
    def get_year_date_started_ended_lines(self, filtered_started_ended_lines, all_year_date_started_ended_lines):
        for filtered in filtered_started_ended_lines:
            tokens_space = filtered.split(' ')
            
            index = 0
            while index < len(tokens_space):
                tokens1 = tokens_space[index].split('T')
                tokens2 = tokens_space[index + 1].split('T')
                
                tokens_comma1 = tokens1[1].split('.')
                tokens_comma2 = tokens2[1].split('.')
                
                tokens_double_quote1 = tokens1[0].split('\"')
                tokens_double_quote2 = tokens2[0].split('\"')
                
                modified_year1 = tokens_double_quote1[1]
                modified_year2 = tokens_double_quote2[1]
                
                all_year_date_started_ended_lines.append(
                    modified_year1 + ' ' + 
                    tokens_comma1[0] + ' ' + 
                    modified_year2 + ' ' + 
                    tokens_comma2[0])
                
                index = index + 2
            
        # For debugging porposes
        #self.print_list(all_year_date_started_ended_lines)
        
    def get_year_month_date_started_ended_lines(self, all_year_date_started_ended_lines, all_year_month_date_started_ended_lines):
        for year_date in all_year_date_started_ended_lines:
            tokens_space = year_date.split(' ')
            
            tokens_dash1= tokens_space[0].split('-')
            tokens_dash2= tokens_space[2].split('-')
            
            tokens_colon1 = tokens_space[1].split(':')
            tokens_colon2 = tokens_space[3].split(':')
            
            all_year_month_date_started_ended_lines.append(
                tokens_dash1[0] + ' ' + tokens_dash1[1] + ' ' + tokens_dash1[2] + ' ' + 
                tokens_colon1[0] + ' ' + tokens_colon1[1] + ' ' + tokens_colon1[2] + ' ' + 
                tokens_dash2[0] + ' ' + tokens_dash2[1] + ' ' + tokens_dash2[2] + ' ' + 
                tokens_colon2[0] + ' ' + tokens_colon2[1] + ' ' + tokens_colon2[2])
        
        # For debugging porposes
        #self.print_list(all_year_month_date_started_ended_lines)
    
    def export_to_excel(self, all_year_month_date_started_ended_lines):
        # Create a workbook
        EXPORTED_FILE_NAME = 'Project4_Group3.xlsx'
        workbook = xlsxwriter.Workbook(EXPORTED_FILE_NAME)
        worksheet = workbook.add_worksheet()
        
        # 12 columns
        columns = [
            'start_year',
            'start_month',
            'start_day',
            'start_hour',
            'start_minute',
            'start_second',
            'end_year',
            'end_month',
            'end_day',
            'end_hour',
            'end_minute',
            'end_second'
        ]
        
        row = 0
        # Write the first column
        for index in range(len(columns)):
            worksheet.write(row, index, columns[index])
            
        row = 1
        column = 0
        for line in all_year_month_date_started_ended_lines:
            tokens = line.split(' ')
            
            # Reset the column
            column = 0
            for token in tokens:
                worksheet.write(row, column, token)
                column = column + 1
                
            row = row + 1
        
        # Close the workbook
        workbook.close()
        
    def print_list(self, list):
        for data in list:
            print(data)
```

```python
def main():
    process_text_data = Process_Text_Data()
    process_text_data.extract_data_into_lists()
    
if __name__ == '__main__':
    main()
```
